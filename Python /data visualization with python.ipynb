```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(42)
scores = np.random.normal(loc=75, scale=10, size=1000)  # Среднее 75, стандартное отклонение 10

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(scores, bins=20, color='mediumslateblue', edgecolor='black', alpha=0.7)

mean_score = np.mean(scores)
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=2, label=f'Среднее = {mean_score:.2f}')

plt.title('Распределение оценок студентов', fontsize=16)
plt.xlabel('Оценка', fontsize=12)
plt.ylabel('Количество студентов', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2,
             patches[i].get_height(),
             int(n[i]),
             ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
months = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']
traffic_data = pd.DataFrame({
    'Direct': np.random.randint(100, 200, 12),
    'Organic Search': np.random.randint(80, 160, 12),
    'Referral': np.random.randint(50, 100, 12),
    'Social Media': np.random.randint(30, 90, 12),
}, index=months)

plt.figure(figsize=(10, 6))
plt.stackplot(traffic_data.index,
              traffic_data['Direct'],
              traffic_data['Organic Search'],
              traffic_data['Referral'],
              traffic_data['Social Media'],
              labels=traffic_data.columns,
              colors=['#4C72B0', '#55A868', '#C44E52', '#8172B2'],
              alpha=0.85)

plt.title('Источники трафика по месяцам', fontsize=16)
plt.xlabel('Месяц')
plt.ylabel('Трафик (тыс. пользователей)')
plt.legend(loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
categories = ['Электроника', 'Одежда', 'Продукты', 'Канцтовары', 'Игрушки']
sales = [150, 90, 220, 75, 110]

plt.figure(figsize=(9, 6))
bars = plt.bar(categories, sales, color='cornflowerblue', edgecolor='black')

plt.title('Продажи по категориям товаров', fontsize=16)
plt.xlabel('Категория', fontsize=12)
plt.ylabel('Продажи (тыс. ед.)', fontsize=12)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 3,
             f'{height}', ha='center', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
labels = ['Продажи', 'Подписки', 'Партнеры', 'Реклама', 'Прочее']
sizes = [40, 25, 15, 10, 10]
colors = ['#4CAF50', '#2196F3', '#FFC107', '#FF5722', '#9E9E9E']
explode = (0.05, 0.05, 0.05, 0.1, 0.1)  

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    textprops={'fontsize': 12, 'color': 'black'}
)

plt.setp(autotexts, size=12, weight='bold', color='white')
plt.title('Источники дохода компании — 2024 г.', fontsize=16)

legend_labels = [f"{label} — {size}%" for label, size in zip(labels, sizes)]
plt.legend(wedges, legend_labels, title="Категории", loc="center left", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

```python
months = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']
users = [1200, 1500, 1700, 2100, 2300, 2500, 3000, 3200, 3500, 3700, 3900, 4200]

df = pd.DataFrame({'Месяц': months, 'Пользователи': users})

plt.figure(figsize=(10, 6))
plt.plot(df['Месяц'], df['Пользователи'], marker='o', linestyle='-', linewidth=2, color='royalblue', label='Активные пользователи')

for i, value in enumerate(df['Пользователи']):
    plt.text(i, value + 100, str(value), ha='center', fontsize=9)

plt.title('Рост числа пользователей по месяцам (2024 г.)', fontsize=16)
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Число пользователей', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
x = np.random.normal(50, 10, 100)
y = 2.5 * x + np.random.normal(0, 20, 100)  
df = pd.DataFrame({'Затраты': x, 'Продажи': y})


plt.figure(figsize=(10, 6))
sns.regplot(x='Затраты', y='Продажи', data=df,
            scatter_kws={"color": "mediumblue", "s": 50, "alpha": 0.6},
            line_kws={"color": "red", "linewidth": 2},
            ci=95)

plt.title('Регрессионный анализ: Затраты vs Продажи', fontsize=16)
plt.xlabel('Затраты на маркетинг (тыс. сом)', fontsize=12)
plt.ylabel('Объём продаж (тыс. ед.)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python

months = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн']
sales = [100, 120, 150, 130, 170, 180]
profit = [20, 25, 22, 30, 35, 40]

fig, ax1 = plt.subplots(figsize=(10, 6))

bars = ax1.bar(months, sales, color='skyblue', label='Продажи')
ax1.set_ylabel('Продажи (тыс. ед.)', fontsize=12)
ax1.set_xlabel('Месяц', fontsize=12)
ax1.tick_params(axis='y')
ax1.set_ylim(0, max(sales) + 50)

ax2 = ax1.twinx()
ax2.plot(months, profit, color='orange', marker='o', linewidth=2, label='Прибыль')
ax2.set_ylabel('Прибыль (тыс. сом)', fontsize=12)
ax2.tick_params(axis='y')
ax2.set_ylim(0, max(profit) + 10)

plt.title('Комбинированная диаграмма: Продажи и Прибыль', fontsize=16)
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval + 3, int(yval), ha='center', fontsize=9)

for i, val in enumerate(profit):
    ax2.text(i, val + 1, f"{val}", ha='center', fontsize=9, color='darkorange')

plt.tight_layout()
plt.show()
```
