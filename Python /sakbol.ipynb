## Ниже будет информация о вероятности наступления страхового случая

Из-за того что у компании нет достаточной информации о случаях аварии бизнес-джетов, я решил брать информацию из открытых источников.
Я взял информация с этого сайта https://dzen.ru/a/ZKwK2h_gmxKUKX18
Тут говрится, что коэффициент аварийности составляет 1,049 на 100 тыс. летных часов. Переводим коэффициент в вероятность на 1 летный час
1,049 / 100000 = 0,00001049 . Это вероятность аварии на 1 летный час.
В заявлении указано: 9 посадок в месяц, это примерно 1 полет в 3–4 дня. Учитывая короткие перелеты для бизнес-авиации, один рейс ~1–1.5 часов нормальная оценка.
Значит за 15 дней: около 4–5 рейсов, допустим 5 рейсов × 1.5 часа ≈ 7.5 часов налета.

Вероятность страхового случая за 7.5 часов налета = 0,00001049 * 7.5 = 0,000078675
В процентах: 0,000078675 * 100 = 0,0078675% ~  0,008%
